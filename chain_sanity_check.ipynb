```python

from getdist import plots, MCSamples
import getdist
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# planck path
planck_path = '/home/grads/data/evan/cosmopower_emcee/notebooks/planck_emulated'
# LSST path
lsst_path = '/home/grads/data/evan/mc_data/fiducial'
# joint path
joint_path = '/home/grads/data/evan/LSST_emulation/joint_data'
```

```python
planck = getdist.mcsamples.loadMCSamples(file_root=planck_path, no_cache=True)
lsst = getdist.mcsamples.loadMCSamples(file_root=lsst_path, no_cache=True)
joint = getdist.mcsamples.loadMCSamples(file_root=joint_path, no_cache=True)
```

```python
names_0 = planck.getParamNames().getRunningNames()
names_1 = lsst.getParamNames().getRunningNames()
names_2 = joint.getParamNames().getRunningNames()

print(names_0)
print(names_1)
print(names_2)
joint.setParamNames(['omegab', 'omegac', 'logA', 'ns', 'H0', 'h', 'tau', 'Aplanck'])

common_params = [param for param in names_0 if param in names_1]

g = plots.get_subplot_plotter()
g.settings.num_plot_contours = 2
g.triangle_plot([planck,lsst,joint],
               filled=False,
               params=common_params)
```
