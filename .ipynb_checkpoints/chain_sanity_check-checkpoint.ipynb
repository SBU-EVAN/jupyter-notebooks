```python
from getdist import plots, MCSamples
import getdist
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# planck path
planck_path = '/home/grads/extra_data/evan/planck_chains/3'
# LSST path
lsst_path = '/home/grads/extra_data/evan/lsst_chains/lsst_3'
# joint path
joint_path = '/home/grads/extra_data/evan/joint_chains/joint_3'
```

```python
planck = getdist.mcsamples.loadMCSamples(file_root=planck_path, no_cache=True)
lsst = getdist.mcsamples.loadMCSamples(file_root=lsst_path, no_cache=True)
joint = getdist.mcsamples.loadMCSamples(file_root=joint_path, no_cache=True)

# forgot to use H0 instead of h in planck
s = planck.samples
s[...,2] = 100*s[...,2]
planck.setSamples(s)
```

```python
names_0 = planck.getParamNames().getRunningNames()
names_1 = lsst.getParamNames().getRunningNames()
names_2 = joint.getParamNames().getRunningNames()

print(names_0)
print(names_1)
print(names_2)
joint.setParamNames(['omegabh2', 'omegach2', 'logA', 'ns', 'H0', 'tau', 'Aplanck'])

common_params = [param for param in names_0 if param in names_1]
print(common_params)

param_limits = {
    'omegabh2': [0.01,0.04],
    'omegach2': [0.06,0.17],
    'H0'      : [55,85],
    'ns'      : [0.88,1.06],
    'logA'    : [2.6,3.6]
}

g = plots.get_subplot_plotter()
g.settings.num_plot_contours = 2
g.triangle_plot([lsst,planck,joint],
               filled=False,
               params=common_params,
               param_limits=param_limits)
```

```python
from numpy import random
planck_path = '/home/grads/extra_data/evan/planck_chains/'
ints = random.randint(0, high=900, size=3, dtype=int)

burnin=3000
thin=10

planck_fiducial = getdist.mcsamples.loadMCSamples(file_root='/home/grads/data/evan/cosmopower_emcee/notebooks/planck_emulated', no_cache=True)
planck1 = getdist.mcsamples.loadMCSamples(file_root=planck_path+str(ints[0]), no_cache=True)
planck2 = getdist.mcsamples.loadMCSamples(file_root=planck_path+str(ints[1]), no_cache=True)
planck3 = getdist.mcsamples.loadMCSamples(file_root=planck_path+str(ints[2]), no_cache=True)

samples0 = planck_fiducial.samples[burnin::thin]
weights0 = planck_fiducial.weights[burnin::thin]
samples0[:,2] = 100*samples0[:,2] # wtf
planck_fiducial.setSamples(samples0,weights=weights0)
planck_fiducial.setParamNames(['omegabh2', 'omegach2', 'H0', 'tau', 'ns', 'logA', 'Aplanck'])

samples1 = planck1.samples[burnin::thin]
weights1 = planck1.weights[burnin::thin]
samples1[:,2] = 100*samples1[:,2] # wtf
planck1.setSamples(samples1,weights=weights1)
planck1.setParamNames(['omegabh2', 'omegach2', 'H0', 'tau', 'ns', 'logA', 'Aplanck'])

samples2 = planck2.samples[burnin::thin]
weights2 = planck2.weights[burnin::thin]
samples2[:,2] = 100*samples2[:,2] # wtf
planck2.setSamples(samples2,weights=weights2)
planck2.setParamNames(['omegabh2', 'omegach2', 'H0', 'tau', 'ns', 'logA', 'Aplanck'])

samples3 = planck3.samples[burnin::thin]
weights3 = planck3.weights[burnin::thin]
samples3[:,2] = 100*samples3[:,2] # wtf
planck3.setSamples(samples3,weights=weights3)
planck3.setParamNames(['omegabh2', 'omegach2', 'H0', 'tau', 'ns', 'logA', 'Aplanck'])
```

```python
# We can also check the training region of parameter space for cosmopower
training_parameters = np.load('/home/grads/extra_data/evan/test_chains/camb_tt_training_params.npz')
print(training_parameters.files)
```

```python
params = []

for key in training_parameters.files:
    arr = training_parameters[key]
    if key=='h':
        arr=100*arr
    params.append(arr)
    
params = np.transpose(np.array(params))
    
chain = getdist.MCSamples(samples=params,names=['omegabh2','omegach2','H0','tau','ns','logA'])
```
